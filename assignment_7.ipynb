```python
# problem 1e)
# use while loop
def roll_simulate():
    list_sums = []
    sums = 0
    while True:
        roll = np.random.randint(1,7)
        sums += roll
        list_sums.append(sums)
        if roll == 6:
            break
    avg_pips = sum(list_sums) / len(list_sums)
    return avg_pips

# roll_simulate()

# now going to simulate 5000 times to get the average 
def bootstrap_roll(num_sims):
    list_sims = []
    for _ in range(num_sims):
        result = roll_simulate()
        list_sims.append(result)
    avg_pips = sum(list_sims) / len(list_sims)
    return avg_pips


until_six = bootstrap_roll(5000)


print(f"The expected value of rolling a dice until you roll a six is {until_six}")
```
